### 2. Explain the distribution of probability of any continuous variable.

```python
#variable for Sleep Time

ST = df['SleepTime']
ST.describe()
```

```python
import numpy as np

uST = ST.unique()

n =0
total = 0

tn = []

for i in uST:
    n += 1
    total = len(ST[ST == i])
    tn.append(total/319795)

lsp = np.array([uST,tn]).transpose()


pST = pd.DataFrame(lsp,columns=['Sleep Time','Probability of Sleep Time'])
pST

plt.figure(figsize=(13,6))
sns.barplot(x='Sleep Time', y='Probability of Sleep Time',data=pST)

    
```

```python
import scipy.stats as ss

#skewness variable
sk = ss.skew(pST, axis=0, bias=True)

if sk[1] > 0:
    print("Positively Skewed", sk[1])

else: 
    print("Negatively Skewed", sk[1])
```

The weight of the graph is on the left, which implies that the data is positively skewed. Measuring the skewness of the data the results were greater than 0, once again showing that the data is positively skewed.

```python
kts = ss.kurtosis(pST, axis=0, bias=True)


if kts[1] > 3:
    print("leptokurtic", kts[1])

else: 
    print("playkurtic", kts[1])
```

This data is heavy-tailed and playkurtic because the kurtosis of the distribution is greater than 3.

### 3. What is the probability that a person who smokes has SkinCancer? 

```python
#Number of people with skin cancer
SC = len(df[df['SkinCancer'] == 'Yes'])

#Number of people who smoke
Smokers = len(df[(df["Smoking"] == 'Yes') & (df["SkinCancer"] == 'Yes')])

#the probability that a person who smokes has Skin Cancer

pSmSc = Smokers/SC
print(pSmSc)
```

The probability that person who smokes has skin cancer is 0.4646366410677756
